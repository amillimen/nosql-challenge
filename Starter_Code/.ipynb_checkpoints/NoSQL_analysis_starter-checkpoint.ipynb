```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygienequery ={'scores.Hygiene':{'$eq': 20}}
results = establishments.find(hygienequery)


# Use count_documents to display the number of documents in the result
countTwenty = establishments.count_documents(hygienequery)


print(f'The number of restaurants with a hygiene score of 20 is ', countTwenty)

# Display the first document in the results using pprint
for i in range (1):
    pprint(results[i])

```

```python
# Convert the result to a Pandas DataFrame
HygieneDF =pd.DataFrame(results)

# Display the number of rows in the DataFrame
row_count = HygieneDF.shape[0]
print(f'The number of rows in this dataframe is: ', row_count)

# Display the first 10 rows of the DataFrame
HygieneDF.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
londonquery ={'LocalAuthorityName':{'$regex':'London'},'RatingValue':{'$gte': 4}}
resultsfour = establishments.find(londonquery)

# Use count_documents to display the number of documents in the result
countLondon = establishments.count_documents({'LocalAuthorityName':{'$regex':'London'},'RatingValue':{'$gte': 4}})

print(f'The number of restaurants with a hygiene score of >=4 is', countLondon)

# Display the first document in the results using pprint
for i in range (1):
    pprint(resultsfour[i])
```

```python
# Convert the result to a Pandas DataFrame
LondonFourDF =pd.DataFrame(resultsfour)

#Count the number of rows in the dataframe
row_countLondon = LondonFourDF.shape[0]
print(f'The number of rows in this dataframe is: ', row_countLondon)

# Display the first 10 rows of the DataFrame
LondonFourDF.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score


#Penange info: 'geocode': {'latitude': '51.49014200', 'longitude': '0.08384000'}
degree_search = 0.01
fields = {'BusinessName', 'scores.Hygiene'}
Penangquery = {
                'geocode.latitude': {'$gte':51.48014200,'$lte':51.50014200},
                'geocode.longitude':{'$gte':.0738400,'$lte':.0938400},
                'RatingValue':{'$eq': 5}
}              
  
sort = [('scores.Hygiene', -1)]

results = list(establishments.find(Penangquery, fields).sort(sort))
print("Number of documents in result:", establishments.count_documents(Penangquery))

pprint(results)
```

```python
# Convert result to Pandas DataFrame
PenangqueryDF = pd.DataFrame(results)
print(len(PenangqueryDF))
PenangqueryDF.head(5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}
               
# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': '$LocalAuthorityName', 
                          'count': { '$sum': 1 },
                         }}
               
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1 }}

#create the pipeline
pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"The number of documents in the result is: ", len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
zeroHygieneDF = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f'The number of rows in the dataframe is:', len(zeroHygieneDF))

# Display the first 10 rows of the DataFrame
zeroHygieneDF.head(10)
```
